```
# get chords get a list of links and scrape chords from them
def getChords(links):
  chords = []
  for link in links: 
    print(link)
    song = requests.get(link)
    song_soup = bs4.BeautifulSoup(song.content, 'html5lib')
    song_chords = []
    for chord in song_soup.find_all('a', title=True, href=False):
      chord = chord.string
      if chord == None or chord.startswith('\xa0') or chord == ' ':
        continue
      song_chords.append(chord)
    chords.append(song_chords)
  return chords
```

### Create One-hot Vectors for Chords to Input into RNN

```
chord_to_index = {}
index_to_chord = {}

for i in range(len(chords)):
  chord_to_index[chords[i]] = i
  index_to_chord[i] = chords[i]

print(chord_to_index)
print(index_to_chord)
```

```
# set global variables
m = len(raw_X) # number of examples
chord_space = len(chords)+1 # number of unique chords, the last chord is empty chord for initialization

X = np.zeros((m, max_song_len, chord_space))
Y = np.zeros((m, max_song_len, chord_space))
print(X.shape)
print(Y.shape)
```

```
index = -1
for i in range(m):
  song = raw_X[i]
  for j in range(len(song)):
    chord = song[j]
    # ith song, jth chord, index
    X[i][j][index] = 1
    index = chord_to_index[chord]
    Y[i][j][index] = 1
  
# print out the data for first song to check
print(raw_X[0])
print(Y[0][0])
print(X[0][1])
print(Y[0][1])
print(X[0][2])

sample_C = Y[0][0]
```

### Train the Model

```
from tensorflow.python.keras.layers import Dense, LSTM
from tensorflow.python.keras import Sequential
from tensorflow.python.keras.callbacks import LambdaCallback
```

```
model = Sequential()
model.add(LSTM(128, input_shape=(max_song_len, chord_space), return_sequences=True))
model.add(Dense(chord_space, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam')
model.fit(X, Y, batch_size=64, epochs=500)
```

```
def chords_to_one_hot(cs):
  one_hot = np.zeros((len(cs),chord_space))
  for i in range(len(cs)):
    c = cs[i]
    index = chord_to_index[c]
    # ith chord, index
    one_hot[i][index] = 1
  return one_hot
```

```
print(chords_to_one_hot(['C', 'G/B']))
```

```
chord_sequence = ['C', 'Am7']
Tx = len(chord_sequence)
x_pred = np.zeros((1, Tx, chord_space))
x_pred[0] = chords_to_one_hot(chord_sequence)
  

prediction = model.predict(x_pred)
print(prediction)

prediction = prediction[0][-1]

# get the top k most probable chords
top_k = 5
max_prob_indices = np.argsort(prediction)[::-1][:top_k]
top_chords = []
for index in max_prob_indices:
  top_chords.append(index_to_chord[index])

print(top_chords)

'''
print(x_pred.shape)
print(prediction.shape)

max_prob = np.max(prediciton, axis=None)
print("max_prob: " + str(max_prob))

print("chord_space: " + str(chord_space))
'''
```
